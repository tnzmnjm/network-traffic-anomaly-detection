```python
import pandas as pd
import joblib
import plotly.express as px
```

```python
df_ip_103 = joblib.load('df_ip_103.csv')
```

```python
df_ip_103 = df_ip_103.sample(1000)
```

```python
df_ip_103.head(5)
```

```python
df_ip_103_time = df_ip_103.groupby(['frame.time','ip.src', 'tcp.srcport', 'tcp.dstport']).size().reset_index(name='count')
```

```python
df_ip_103['frame.time'] = pd.to_datetime(df_ip_103['frame.time'])
```

```python
df_ip_103.head(3)
```

```python
fig = px.scatter(df_ip_103_time,
                 x='tcp.srcport',
                 y='tcp.dstport',
                 size='count',
                 color='frame.time',  # This adds the time dimension
                 hover_data=['ip.src', 'count', 'frame.time'],
                 title='Connections Over Time for IP 10.20.30.103',
                 color_continuous_scale=px.colors.sequential.Viridis)

# Update layout for better readability
fig.update_layout(xaxis_title='Source Port', yaxis_title='Destination Port',
                  xaxis=dict(type='category'), yaxis=dict(type='category'))

# Show the plot
fig.show()
```
